# Applying ML algorithms (training)

## - Training

```python
# Separating the data
x = np.array(passengers_data.drop(["Survived"], 1)) # Data of passengers
y = np.array(passengers_data["Survived"]) # Survived or not

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# K means clustering
kmeans = KMeans(n_clusters=5).fit(X=x_train, y=y_train)
centroids = kmeans.cluster_centers_
print(centroids)
kmeans
```

```python
kmeans.score(X=x_test, y=y_test)
#plt.scatter(x, y)
#plt.show()
```

#### - Linear regression

```python
# Linear regression
reg = LinearRegression().fit(X=x_train, y=y_train)
reg.score(X=x_train, y=y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x_train, y_train)
r_value ** 2 # Fit of the model, if there is linear fit.
```

```python
reg.predict([[2, 0, '5', 2, 1, 34.7, 1]]) 
```

#### - Logistic regression

```python
log_reg = LogisticRegression(random_state=0).fit(X=x_train, y=y_train)
log_reg.score(X=x_train, y=y_train)
```

```python
log_reg.predict(x_test)
```

```python
log_reg.predict_proba(x_test)
```

#### - SVM

```python
clf = svm.SVC().fit(X=x_train, y=y_train)
clf.score(X=x_train, y=y_train)
```

```python
clf.predict(x_test)
```

#### - Random Forest

```python
X, y = make_classification(n_samples=1000, n_features=4,
                           n_informative=2, n_redundant=0,
                           random_state=0, shuffle=False)

rndF = RandomForestClassifier(max_depth=10, random_state=0)
rndF.fit(X=x_train, y=y_train)
rndF.score(X=x_train, y=y_train)
```
